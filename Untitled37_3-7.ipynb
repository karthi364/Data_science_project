```python
df['BMI'] = df['weight'] / ((df['height']/100) ** 2)
```

```python
df['age'] = (df['age'] / 365).astype(int)
```

```python
df['pulse_pressure'] = df['ap_hi'] - df['ap_lo']
df['mean_arterial_pressure'] = df['ap_lo'] + (df['pulse_pressure'] / 3)
df['bp_ratio'] = df['ap_hi'] / df['ap_lo']
```

```python
df['BMI_category'] = pd.cut(
    df['BMI'],
    bins=[0,18.5,25,30,100],
    labels=[0,1,2,3]   # underweight, normal, overweight, obese
)
```

```python
df['age_group'] = pd.cut(
    df['age'],
    bins=[30,40,50,60,70],
    labels=[1,2,3,4]
)
```

```python
df['risk_score'] = (
    df['cholesterol'] +
    df['gluc'] +
    df['smoke'] +
    df['alco'] +
    (1 - df['active'])
)
```

```python
df['hypertension'] = (
    (df['ap_hi'] >= 140) | (df['ap_lo'] >= 90)
).astype(int)
```

```python
df['age_bmi'] = df['age'] * df['BMI']
df['bp_age'] = df['ap_hi'] * df['age']
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)
```

```python
print(df.head())
```

```python
# Gender
df['gender_label'] = df['gender'].map({1: 'Female', 2: 'Male'})

# Cholesterol
df['cholesterol_label'] = df['cholesterol'].map({
    1: 'Normal',
    2: 'Above Normal',
    3: 'Well Above Normal'
})

# Glucose
df['glucose_label'] = df['gluc'].map({
    1: 'Normal',
    2: 'Above Normal',
    3: 'Well Above Normal'
})

# Smoking
df['smoke_label'] = df['smoke'].map({
    0: 'Non-Smoker',
    1: 'Smoker'
})

# Alcohol
df['alco_label'] = df['alco'].map({
    0: 'No Alcohol',
    1: 'Drinks Alcohol'
})

# Physical Activity
df['active_label'] = df['active'].map({
    0: 'Inactive',
    1: 'Active'
})

# Heart Disease
df['Heart_label'] = df['Heart_Disease'].map({
    0: 'No Disease',
    1: 'Heart Disease'
})
```

```python

df = df[(df['ap_hi'] >= 70) & (df['ap_hi'] <= 200)]
df = df[(df['ap_lo'] >= 40) & (df['ap_lo'] <= 130)]
df = df[df['ap_hi'] > df['ap_lo']]
df = df[(df['height'] >= 140) & (df['height'] <= 210)]
df = df[(df['weight'] >= 40) & (df['weight'] <= 180)]


df = df[(df['BMI'] >= 10) & (df['BMI'] <= 60)]
print(f"Rows after cleaning: {len(df)}")
```

```python

X = df[['age', 'gender', 'height', 'weight', 'BMI',
        'ap_hi', 'ap_lo',
        'pulse_pressure', 'mean_arterial_pressure',
        'cholesterol', 'gluc', 'smoke', 'alco', 'active',
        'BMI_category', 'age_group', 'risk_score', 'hypertension']]

X = X.copy()
X['BMI_category'] = X['BMI_category'].astype(int)
X['age_group'] = X['age_group'].astype(int)

y = df['Heart_Disease']
```

```python
X['BMI_category'] = X['BMI_category'].astype(int)
X['age_group'] = X['age_group'].astype(int)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,stratify=y,test_size=0.3, random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

import matplotlib.pyplot as plt
```

i


```python
lr = LogisticRegression(max_iter=1000,class_weight='balanced')

param_grid_lr = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['lbfgs', 'liblinear']
}

grid_lr = GridSearchCV(lr, param_grid_lr, cv=5, scoring='accuracy', n_jobs=-1)
grid_lr.fit(X_train, y_train)

best_lr = grid_lr.best_estimator_

y_pred_lr = best_lr.predict(X_test)

print("Best LR Params:", grid_lr.best_params_)
print(classification_report(y_test, y_pred_lr))
```

```python
rf = RandomForestClassifier()

param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

grid_rf = GridSearchCV(rf, param_grid_rf, cv=3, scoring='accuracy', n_jobs=-1)
grid_rf.fit(X_train, y_train)

best_rf = grid_rf.best_estimator_

y_pred_rf = best_rf.predict(X_test)

print("Best RF Params:", grid_rf.best_params_)
print(classification_report(y_test, y_pred_rf))
```

```python
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss')

param_grid_xgb = {
    'n_estimators': [100, 200],
    'max_depth': [3, 6],
    'learning_rate': [0.01, 0.1],
    'subsample': [0.8, 1]
}

grid_xgb = GridSearchCV(xgb, param_grid_xgb, cv=3, scoring='accuracy', n_jobs=-1)
grid_xgb.fit(X_train, y_train)

best_xgb = grid_xgb.best_estimator_

y_pred_xgb = best_xgb.predict(X_test)

print("Best XGB Params:", grid_xgb.best_params_)
print(classification_report(y_test, y_pred_xgb))
```

```python
import lightgbm as lgb
from lightgbm import LGBMClassifier
```

```python
lgbm = LGBMClassifier(class_weight='balanced', random_state=42)

param_grid_lgbm = {
    'n_estimators': [100, 200],
    'num_leaves': [31, 50],
    'learning_rate': [0.05, 0.1],
    'max_depth': [-1, 10]
}

grid_lgbm = GridSearchCV(lgbm, param_grid_lgbm, cv=3, scoring='accuracy', n_jobs=-1)
grid_lgbm.fit(X_train, y_train)

best_lgbm = grid_lgbm.best_estimator_

y_pred_lgbm = best_lgbm.predict(X_test)

print("Best LightGBM Params:", grid_lgbm.best_params_)
print(classification_report(y_test, y_pred_lgbm))
```

```python
models = {
    "Logistic Regression": y_pred_lr,
    "Random Forest": y_pred_rf,
    "XGBoost": y_pred_xgb,
    "LightGBM": y_pred_lgbm
}

for name, y_pred in models.items():
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)

    disp.plot()
    plt.title(name)
    plt.show()
```

```python
from sklearn.metrics import roc_curve, auc, RocCurveDisplay
```

```python
plt.figure(figsize=(8,6))

models_prob = {
    "Logistic Regression": best_lr,
    "Random Forest": best_rf,
    "XGBoost": best_xgb,
    "LightGBM": best_lgbm
}

for name, model in models_prob.items():
    # Get predicted probabilities for the positive class
    y_prob = model.predict_proba(X_test)[:, 1]

    fpr, tpr, thresholds = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

# Plot random chance line
plt.plot([0, 1], [0, 1], 'k--', label='Random Chance')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves')
plt.legend(loc='lower right')
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import pandas as pd
```

```python
# Collect models and their predicted probabilities
models = {
    "Logistic Regression": best_lr,
    "Random Forest": best_rf,
    "XGBoost": best_xgb,
    "LightGBM": best_lgbm
}

metrics = []

for name, model in models.items():
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    metrics.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1-score": f1_score(y_test, y_pred),
        "AUC": roc_auc_score(y_test, y_prob)
    })

# Create a DataFrame
metrics_df = pd.DataFrame(metrics).sort_values(by='AUC', ascending=False)
metrics_df
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Metrics we want to plot
metrics_to_plot = ["Accuracy", "F1-score", "AUC"]

# Set bar width and positions
x = np.arange(len(metrics_to_plot))
width = 0.2  # width of each bar

# Colors for each model
colors = ['skyblue', 'lightgreen', 'salmon', 'plum']
```

```python
plt.figure(figsize=(10,6))

for i, model in enumerate(metrics_df['Model']):
    values = metrics_df.loc[metrics_df['Model'] == model, metrics_to_plot].values.flatten()
    plt.bar(x + i*width, values, width=width, label=model, color=colors[i])

# X-axis labels
plt.xticks(x + width*1.5, metrics_to_plot)

plt.ylabel("Score")
plt.ylim(0, 1)
plt.title("Model Comparison: Accuracy, F1-score, and AUC")
plt.legend()
plt.show()
```
